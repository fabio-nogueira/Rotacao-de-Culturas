```python
import numpy as np
from itertools import permutations
from random import *
import math
import random
import pandas as pd
```

```python
Planilha = pd.ExcelFile('RotacaoCulturas.xlsx')
TabelaCulturas = pd.read_excel(Planilha,  'Culturas')
TabelaLucratividade = pd.read_excel(Planilha,  'Lucratividade')
TabelaProdutividade = pd.read_excel(Planilha,  'Produtividade')
```

```python
TabelaCulturas
```

```python
L = 16 #Numero maximo de lotes definido no artigo
M = 24 #Numero de periodos de plantio definido no artigo (2 anos)
```

```python
CulturaIndice = TabelaCulturas.CulturaIndice.to_numpy() #Transformando a Coluna da Planilha em Numpy Array
FamiliaIndice = TabelaCulturas.FamiliaIndice.to_numpy() #Transformando a Coluna da Planilha em Numpy Array
CicloEmTi = TabelaCulturas.CicloEmTi.to_numpy() #Transformando a Coluna da Planilha em Numpy Array

Plantas = CulturaIndice[-1] #Ultimo elemento do indice de Culturas é o número total de Plantas
Familias = FamiliaIndice[-1] #Ultimo elemento do indice de familias é o número total de Famílias
print("Numero de Plantas + Pousio: ", Plantas)
print("Numero de Familias: ", Familias)
vetorPlantasSorteado = np.array(CulturaIndice) #um vetor pra sortear os indices futuramente
```

```python
def sorteio(l):
    random.shuffle(l)
    print("Vetor de Plantas sorteado: ", l)
    return l
```

```python
sorteio(vetorPlantasSorteado) #Realiza um sorteio do vetor de Plantas, ou seja, Escolhe uma permutação e
                                #salva no mesmo vetorPlantasSorteado
```

```python
def FamiliaConsecutiva(): #Analisa se no vetor sorteado, plantas da mesma familia estao sendo plantadas consecutivamente
                            #Se sim, o vetor é sorteado novamente.
    for i in range(Plantas):
        x2=0
        y2=0
        while (x2==y2): #Começa sendo valido, só sai quando são familias diferentes
            if (i+1==Plantas): #se o vetor terminou
                break
            x = vetorPlantasSorteado[i]  #PlantaAtual
            y = vetorPlantasSorteado[i+1] #ProximaPlanta
            x2 = FamiliaIndice[x-1] #Familia da Planta Atual
            y2 = FamiliaIndice[y-1] #Familia da Próxima Planta
            if (x2==y2):
                if(x2==11): #Se for pousio, pode repetir
                    break
                print("A Planta ", vetorPlantasSorteado[i], "Pertence a Familia", FamiliaIndice[x-1], "E a Planta ", vetorPlantasSorteado[i+1], "Tambem pertence a Familia ", FamiliaIndice[y-1] )
                print("É preciso sortear novamente")
                sorteio(vetorPlantasSorteado) #sorteia
                i=0 #volta pro começo
            
```

```python
FamiliaConsecutiva()
```

```python
#Contador de repeticoes, faz um vetor contando a quantidade de vezes que uma planta eh plantada do vetorPlantasSorteado
def contador():
    c=[]
    for i in range(Plantas):
        x=vetorPlantasSorteado[i]
        y=CicloEmTi[x-1]
        print("Planta sorteada %d, Quantidade de vezes que ela é plantada: %d" % (x, y))
        c.append(y)
    contador=np.array(c)
    return contador
```

```python
contador()
```

```python
#Obtendo o construtor:
def construtor():
    teste = 0 
    while (teste == 0):
        sorteio(vetorPlantasSorteado) #Realiza sorteio
        FamiliaConsecutiva() #Verifica se tem familia Consecutiva no mesmo lote, se tiver sorteia ate nao ter
        c = contador() #Faz o contador do sorteio
        construtor = np.repeat(vetorPlantasSorteado, c)
        if (construtor[M-1] != construtor[M]):  #se o tempo de plantacao cabe na linha
            #print("Vetor Final: ",construtor)
            teste = 1
    return construtor    
```

```python
construtor()
```

```python
#Agora é só repetir em todos os Lotes, de 0 até L
Cult = np.zeros((L, M))
Families = np.zeros((L, M))
constr = construtor()
for i in range(L):
    for j in range(M):
        Cult[i,j]=constr[j]
    constr = construtor()
```

```python
Cult = Cult.astype(int)
```

```python
Cult
```

```python
for i in range(L):
    for j in range(M):
        x=Cult[i,j]
        Families[i,j]=FamiliaIndice[x-1]
        
```

```python
CultFinal1 = pd.DataFrame(Cult, columns=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro', 'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])
CultFinal2 = pd.DataFrame(Families, columns=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro', 'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])
```
